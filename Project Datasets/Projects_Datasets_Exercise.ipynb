```python
#1 SALES DATA ANALYSIS & VISUALISATION

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('superstore_final_dataset.csv')
print(data.head(5),data.dtypes,data.describe())

print('\n\n')

print("3---MONTHLY SALES BY TREND---")
data = pd.DataFrame({
    'Order Date': ['08/11/2017','08/11/2017','12/06/2017','11/10/2016','11/10/2016'],
    'Sales': [261.9600,731.9400,14.6200,957.5775,22.3680]
})

data['Order Date'] = pd.to_datetime(data['Order Date'])
data['month'] = data['Order Date'].dt.month
monthly_sales = data.groupby('month')['Sales'].sum()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o',linestyle='--')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

print('\n\n')

print("4---TOP TEN PRODUCTS BY SALES---")

data = pd.DataFrame({
    'Category': ['Furniture', 'Furniture', 'Office Supplies', 'Furniture', 'Office Supplies','Furniture','Office Supplies','Technology','Office Supplies','Office Supplies'],
    'Sales': [261.96, 731.94, 14.62, 957.57, 22.37,48.86,7.28,907.152,18.50,114.9]
})

groupedData = data.groupby('Category')['Sales'].sum().reset_index()
sortedData = groupedData.sort_values('Sales', ascending=False)

plt.bar(sortedData['Category'], sortedData['Sales'], color='purple')
plt.title('Top Products Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()

print('\n\n')

print("5---REGIONAL SALES DISTRIBUTION---")

data = pd.DataFrame({
    'Region': ['South', 'West','Central','East','South'],
    'Sales': [261.9600,731.9400,14.6200,957.5775,22.3680]
})

plt.figure(figsize=(6, 8))
plt.pie(data['Sales'], labels=data['Region'], autopct='%1.1f%%', startangle=90, colors=['blue', 'purple', 'orange','Violet','green'])
plt.title('Pie Chart Showing Regional Sales')
plt.axis('equal')  
plt.show()
```

```python
# 2 CUSTOMER SEGMANTATION(AGE VS SPENDING SCORE)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Mall_Customers segmentation data.csv')
print(df.isnull().sum())
print('\n')

#Calculate Mean,Median, and  Quartiles for Age and Spending Scores
meanAge = df['Age'].mean()
meanSpendingScore = df['Spending Score (1-100)'].mean()

medianAge = df['Age'].median()
medianSpendingScore = df['Spending Score (1-100)'].median()

q1Age = df['Age'].quantile(0.25)
q3Age = df['Age'].quantile(0.75)
q1SpendingScore = df['Spending Score (1-100)'].quantile(0.25)
q3SpendingScore = df['Spending Score (1-100)'].quantile(0.75)

print('Calculating Mean for Age and Spending Score:')
print('Mean Age:', meanAge)
print('Mean Spending Score:',meanSpendingScore)

print('\nCalculating Median for Age and Spending Score:')
print('Meadian Age:',medianAge)
print('Median Spending Score:',medianSpendingScore)

print('\nCalculating Quartiles for Age and Spending Score:')
print('Age Quartiles for Quartile1:',q1Age, ',and For Quartile3:',q3Age)
print('Spending Score Quartiles for Quartile1:',q1SpendingScore, 'and For Quartile3:',q3SpendingScore)

# Scatter Plot Analysis
print('\n')
plt.figure(figsize = (8,5))
plt.scatter(df['Age'],df['Spending Score (1-100)'],color='Purple', alpha=0.7)
plt.title('Scatter Plot Analysis',fontsize=16)
plt.xlabel('Age',fontsize=10)
plt.ylabel('Spending Score (1-100)',fontsize=10)
plt.grid(True,alpha=0.3)
plt.show()

# Histogram of Spending Distribution Across the Age Groups
print('\n')
plt.figure(figsize = (10,8))
plt.hist(df['Age'],bins = 10,color='Purple', alpha=0.7)
plt.title('Histogram of Spending Distribution Across Age Groups',fontsize=16)
plt.xlabel('Age',fontsize=10)
plt.ylabel('Spending Score (1-100)',fontsize=10)
# plt.grid(True,alpha=0.3)
plt.show()
```
